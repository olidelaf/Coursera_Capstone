```python
import pandas as pd
import requests
import json
```

```python
# Upload data relating to house prices, postcodes, latitude and longitude. Merge the two dataframes
pp = pd.read_csv("pp-2018.csv") 
del pp['Neighbourhood']

# remove NaNs
pp.dropna(inplace = True) 

# upload postcodes data
postcodes = pd.read_csv("postcodes.csv")

print(len(pp))
```

```python
# extract first letter form second half of post code
mylist = postcodes.Postcode.str.split(' ',1).str[1]       
firstletter = [x[0] for x in mylist]
```

Delete the secondary half of the postcode so that we can keep the bit of postcode which relates to the larger area

```python
# Create new postcode variable with just one letter from second half

# dataframe of just post code split in half
df = pd.DataFrame(postcodes.Postcode.str.split(' ',1).tolist(), columns = ['Postcode','Postcode 2'])

# Format new postcode
pcarea = df['Postcode'] + ' ' +firstletter
```

Merge the new postcode data into the old postcode data

```python
# Attach new postcode format
postcodes['Postcode'] = pcarea

# Postcode amendment complete
postcodes = postcodes.groupby('Postcode', as_index=False)['Latitude','Longitude'].mean()
```

```python
# Clean pp postcodes

mylist2 = pp.Postcode.str.split(' ',1).str[1]
firstletter2 = [x[0] for x in mylist2]

df2 = pd.DataFrame(pp.Postcode.str.split(' ',1).tolist(), columns = ['Postcode','Postcode 2'])
pparea = df2['Postcode'] + ' ' +firstletter2
```

```python
# Update newly formatted postcodes to original dataframe
pp = pp.reset_index()
pp['Postcode'] = pparea
```

```python
def getNearbyVenues(names, latitudes, longitudes, radius=500):
    count = 0
    venues_list=[]
    for name, lat, lng in zip(names, latitudes, longitudes):
            
        # create the API request URL
        url = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}'.format(
            CLIENT_ID, 
            CLIENT_SECRET, 
            VERSION, 
            lat, 
            lng, 
            radius, 
            LIMIT)
        
        # make the GET request
        results = requests.get(url).json()["response"]['groups'][0]['items']
        
        # return only relevant information for each nearby venue
        venues_list.append([(
            name, 
            lat, 
            lng, 
            v['venue']['name'], 
            v['venue']['location']['lat'], 
            v['venue']['location']['lng'],  
            v['venue']['categories'][0]['name']) for v in results])

    nearby_venues = pd.DataFrame([item for venue_list in venues_list for item in venue_list])
    nearby_venues.columns = ['Neighbourhood', 
                  'Neighbourhood Latitude', 
                  'Neighbourhood Longitude', 
                  'Venue', 
                  'Venue Latitude', 
                  'Venue Longitude', 
                  'Venue Category']
    
    return(nearby_venues)
```
