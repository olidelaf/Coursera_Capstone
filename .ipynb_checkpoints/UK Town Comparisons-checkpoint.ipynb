```python
#We will now check how many unique venue category types there are
print('There are {} uniques categories of venue locations.'.format(len(allvenues['Venue Category'].unique())))
```

```python
#Let's consider how many of each type of venue category is in each neighbourhood on average
```

```python
# one hot encoding
uk_onehot = pd.get_dummies(allvenues[['Venue Category']], prefix="", prefix_sep="")

# add neighborhood column back to dataframe
uk_onehot['Neighbourhood'] = allvenues['Neighbourhood'] 

# move neighborhood column to the first column
fixed_columns = [uk_onehot.columns[-1]] + list(uk_onehot.columns[:-1])
uk_onehot = uk_onehot[fixed_columns]

uk_grouped = uk_onehot.groupby('Neighbourhood').mean().reset_index()
uk_grouped.head()
```

```python
#What are the top venues for each neighbourhood
num_top_venues = 10

for hood in uk_grouped['Neighbourhood']:
    print("----"+hood+"----")
    temp = uk_grouped[uk_grouped['Neighbourhood'] == hood].T.reset_index()
    temp.columns = ['venue','freq']
    temp = temp.iloc[1:]
    temp['freq'] = temp['freq'].astype(float)
    temp = temp.round({'freq': 2})
    print(temp.sort_values('freq', ascending=False).reset_index(drop=True).head(num_top_venues))
    print('\n')
```

```python
#We now organise this in order
def return_most_common_venues(row, num_top_venues):
    row_categories = row.iloc[1:]
    row_categories_sorted = row_categories.sort_values(ascending=False)
    
    return row_categories_sorted.index.values[0:num_top_venues]
```

```python
fakeset = allvenues[0:5]
fakeukgroup = uk_grouped[0:5]
```

```python

num_top_venues = 10

indicators = ['st', 'nd', 'rd']

columnstest = ['Neighbourhood']
for ind in np.arange(num_top_venues):
    try:
        columnstest.append('{}{} Most Common Venue'.format(ind+1, indicators[ind]))
    except:
        columnstest.append('{}th Most Common Venue'.format(ind+1))
```

```python
neighbourhoods_venues_sorted = pd.DataFrame(columns=columnstest)
neighbourhoods_venues_sorted['Neighbourhood'] = fakeukgroup['Neighbourhood']

for ind in np.arange(fakeukgroup.shape[0]):
    neighbourhoods_venues_sorted.iloc[ind, 1:] = return_most_common_venues(fakeukgroup.iloc[ind, :], num_top_venues)
```

# Machine Learning: K - Means Clustering

```python
from sklearn.cluster import KMeans
clustering = allvenues.drop(['Neighbourhood','Venue', 'Venue Category','Venue Latitude','Venue Longitude'], 1)
kmeans = KMeans(n_clusters=5, random_state=0).fit(clustering)
labels = kmeans.predict(clustering)
```

```python
labels
```

```python
allvenues = allvenues.assign(Cluster = labels) 
centroids = kmeans.cluster_centers_
```
